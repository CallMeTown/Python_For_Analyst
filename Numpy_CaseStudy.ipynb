## Doc Exercise

```python
import numpy as np

# Random Values 
np.random.seed(0)
sale_amounts = np.random.randint(20, size = [5,3])
```

```python
sale_amounts
```

```python
# Insert index and column
import pandas as pd
quarterly_sales = pd.DataFrame(sale_amounts, index = ['Sugar', 'Salt', 'Cheese', 'Oil', 'Rice'],
                           columns = ['Jan', 'Junly', 'Sept'])
```

```python
quarterly_sales
```

```python
# Get price
prices = np.array([10, 8, 12])
```

```python
data_price = pd.DataFrame(prices.reshape([1,3]), index = ['Price'],
                            columns = ['Jan', 'Junly', 'Sept'])
```

```python
data_price
```

```python
quarterly_sales.shape, data_price.shape
```

```python
# doc quarterly sales with data price
total_price = quarterly_sales.dot(data_price.T)
total_price
```

```python
quarterly_sales.shape, data_price.T.shape
```

```python
# Add a column Total price
quarterly_sales['Total_prices'] = total_price
quarterly_sales
```
